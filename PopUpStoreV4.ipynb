```python
def get_df():
    url = 'https://api.fotocasa.es/PropertySearch/Search?combinedLocationIds=724,14,28,173,0,28079,0,0,0&culture=es-ES&featureIds=130&hrefLangCultures=ca-ES%3Bes-ES%3Bde-DE%3Ben-GB&isMap=false&isNewConstruction=false&isNewConstructionPromotions=false&latitude=40.4096&longitude=-3.68624&pageNumber=1&platformId=1&sortOrderDesc=true&sortType=bumpdate&transactionTypeId=3&propertyTypeId=2'
    resp = r.get(url, headers=headers)
    data = resp.json()
    df = pd.json_normalize(data['realEstates'])
    number_of_iteration=0
    time.sleep(2)
    for i in range(2,80):
        number_of_iteration+=1
        url = f'https://api.fotocasa.es/PropertySearch/Search?combinedLocationIds=724,14,28,173,0,28079,0,0,0&culture=es-ES&featureIds=130&hrefLangCultures=ca-ES%3Bes-ES%3Bde-DE%3Ben-GB&isMap=false&isNewConstruction=false&isNewConstructionPromotions=false&latitude=40.4096&longitude=-3.68624&pageNumber={i}&platformId=1&sortOrderDesc=true&sortType=bumpdate&transactionTypeId=3&propertyTypeId=2'
        resp = r.get(url, headers=headers)
        data = resp.json()
        df = df.append(pd.json_normalize(data['realEstates']))
        if number_of_iteration%10 and number_of_iteration:
            time.sleep(5)
        time.sleep(2)
        print(f'Page number: {i} scanned')
    df.to_json('df.json', orient='records')
    return df
```
